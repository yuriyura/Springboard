# Remove Outliers by using IsolationForest.

```python
df = pd.get_dummies(df, columns=['fuel','title_status','transmission','drive','VClass','region','division','manufacturer'])
#feature_columns = list(train_data.columns.difference(['price']))
```

```python
%%time
from sklearn.ensemble import IsolationForest
clf=IsolationForest(n_estimators=50, max_samples=50, contamination=float(0.004), 
                    max_features=1.0, bootstrap=False, n_jobs=-1,
                    random_state=42, verbose=0,behaviour="new")
# 50 n_nodes, max 50 samples
# 0.04% outlier.
clf.fit(df)
pred = clf.predict(df)
df['anomaly']=pred
outliers=df.loc[df['anomaly']==-1]
outlier_index=list(outliers.index)
#print(outlier_index)
#Find the number of anomalies and normal points here points classified -1 are anomalous
print(df['anomaly'].value_counts())
```

```python
df['anomaly'].value_counts().plot.barh(title='value counts of anomaly');
plt.savefig('compare_normal_vs_anomaly.png');
```
